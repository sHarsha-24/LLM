```python
# Install
# pip install nltk spacy sklearn
# python -m spacy download en_core_web_sm

import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import spacy

# Sample movie reviews
reviews = [
    "The movie Avatar was amazing",
    "I hated the film, worst acting"
]
labels = [1, 0]  # 1=Positive, 0=Negative

# Text Classification
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews)
model = LogisticRegression()
model.fit(X, labels)

# Predict
test_review = ["Avatar has great visuals"]
prediction = model.predict(vectorizer.transform(test_review))

# Named Entity Recognition
nlp = spacy.load("en_core_web_sm")
doc = nlp(test_review[0])
entities = [(ent.text, ent.label_) for ent in doc.ents]

print("Sentiment:", "Positive" if prediction[0] else "Negative")
print("Entities:", entities)
```

```python
# Install
# pip install transformers torch

from transformers import pipeline

qa = pipeline("question-answering")

context = "The movie Avatar was directed by James Cameron and released in 2009."
question = "Who directed Avatar?"

result = qa(question=question, context=context)
print("Answer:", result["answer"])
```

```python
# Install
# pip install transformers torch

from transformers import pipeline

summarizer = pipeline("summarization")

review = """
The movie Avatar has stunning visuals and strong performances.
The story is simple but engaging and the direction is excellent.
Overall, it is a visually immersive experience.
"""

summary = summarizer(review, max_length=30, min_length=10, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```
