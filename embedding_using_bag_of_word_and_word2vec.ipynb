```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["Data science is fun", "Data science is powerful"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

target_word = "science"
index = vectorizer.vocabulary_[target_word]
embedding = X.toarray()[:, index]

print(embedding)
```
